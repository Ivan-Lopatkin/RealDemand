## 📖 Контекст задачи

Вы работаете в небольшой компании по аренде самокатов в Москве.  
Ваша задача — **предсказывать количество поездок в день**, чтобы заранее знать, сколько самокатов вывезти на станции.

У вас есть данные за последние несколько месяцев. Для каждого дня известны:
- 🌡️ **температура воздуха** (в °C)
- 💨 **скорость ветра** (в м/с)

## ⚙️ Подготовка: импорты и генерация данных

Запустите эту ячейку — она создаёт наш датасет и ничего от вас не требует.

Единственная незнакомая вещь здесь — библиотека `numpy`. Пока достаточно знать одно: она умеет работать с **массивами чисел** (как списки в Python, только быстрее и удобнее для математики). Мы будем обращаться с ними так же, как со списками.

```python
import numpy as np
import matplotlib.pyplot as plt

# Фиксируем случайность, чтобы у всех были одинаковые данные
np.random.seed(42)

N = 120  # 120 дней наблюдений

# Генерируем признаки
temperature = np.random.uniform(5, 35, N)   # температура от +5 до +35 °C
wind_speed  = np.random.uniform(0, 10, N)   # ветер от 0 до 10 м/с

# «Настоящая» формула спроса (мы её не знаем — надо угадать!) + шум
true_w_temp = 15.0
true_w_wind = -20.0
true_bias   = 100.0

noise = np.random.normal(0, 30, N)
rides = true_w_temp * temperature + true_w_wind * wind_speed + true_bias + noise
rides = np.maximum(rides, 0).astype(int)  # поездок не может быть отрицательное число

print(f"Датасет создан: {N} дней наблюдений")
print(f"Поездок в день: от {rides.min()} до {rides.max()}, среднее {rides.mean():.0f}")
print(f"Температура: от {temperature.min():.1f} до {temperature.max():.1f} °C")
print(f"Ветер: от {wind_speed.min():.1f} до {wind_speed.max():.1f} м/с")
```

Мы получили список поездок `rides`. Так как это просто список, мы можем делать с ним все привычные преобразования

```python
print(len(rides))
print(rides[:15])
```

## Задача 1. Линейная модель вручную · *1 балл*

### Что такое линейная модель?

На лекции мы узнали, что алгоритм $a(x)$ — это функция, которая по объекту $x$ выдаёт предсказание.

Простейший вид такой функции — **линейная модель**:

$$a(x) = w_1 \cdot x_1 + w_2 \cdot x_2 + b$$

В нашем случае:
- $x_1$ — температура
- $x_2$ — скорость ветра
- $w_1$, $w_2$ — **веса** (коэффициенты, которые мы подбираем)
- $b$ — **сдвиг** (bias), базовый уровень при нулевых признаках

**Пример:** если $w_1 = 10$, $w_2 = -5$, $b = 50$, и сегодня 20°C и ветер 3 м/с, то предсказание:
$$a = 10 \cdot 20 + (-5) \cdot 3 + 50 = 200 - 15 + 50 = 235 \text{ поездок}$$

---

### 2.1 Реализуйте функцию предсказания

Напишите функцию `predict(temperature, wind_speed, w_temp, w_wind, bias)`, которая возвращает предсказанное количество поездок.

```python
def predict(temperature, wind_speed, w_temp, w_wind, bias):
    """
    Линейная модель: предсказывает количество поездок.

    temperature  -- температура в °C (число или массив)
    wind_speed   -- скорость ветра в м/с (число или массив)
    w_temp       -- вес температуры
    w_wind       -- вес ветра
    bias         -- сдвиг
    """
    return w_temp*temperature + wind_speed*w_wind + bias


# Проверка: 10*20 + (-5)*3 + 50 = 235
test_result = predict(20, 3, 10, -5, 50)
print(f"predict(20, 3, 10, -5, 50) = {test_result}")
print(f"Ожидается: 235")
assert test_result == 235, "Что-то не так, перепроверьте формулу!"
print("✅ Тест пройден!")
```

### 2.2 Первые предсказания «на глаз»

Поставьте любые коэффициенты и посмотрите, что модель выдаёт для нескольких конкретных дней.

Пока не думайте о точности — просто убедитесь, что предсказания выглядят разумно (не отрицательные, не миллионы).

```python
# 🔧 ВАШИ КОЭФФИЦИЕНТЫ — попробуйте разные значения!
my_w_temp = 10.0
my_w_wind = -30.0
my_bias   = 300.0

# Смотрим на первые 10 дней: что предсказывает модель vs реальность
print(f"{'День':>5}  {'Темп':>6}  {'Ветер':>6}  {'Реально':>9}  {'Модель':>8}")
print("-" * 45)
for i in range(10):
    pred = predict(temperature[i], wind_speed[i], my_w_temp, my_w_wind, my_bias)
    print(f"{i+1:>5}  {temperature[i]:>5.1f}°  {wind_speed[i]:>5.1f}м/с  {rides[i]:>9}  {pred:>8.0f}")
```

## Задача 2. Функция потерь · *2 балла*

### Как понять, хороша ли модель?

Смотреть на 10 строк вручную — неудобно. Нужно одно число, которое скажет: «модель ошибается в среднем на X поездок».

Это и есть **функция потерь (loss)**. Одна из самых простых — **MAE (Mean Absolute Error)**:

$$\text{MAE} = \frac{1}{n} \sum_{i=1}^{n} |\text{предсказание}_i - \text{реальность}_i|$$

Мы берём ошибку по каждому дню (по модулю, чтобы плюс и минус не гасили друг друга) и усредняем.  
MAE = 50 означает: «в среднем мы ошибаемся на 50 поездок».

---

### 3.1 Реализуйте MAE

**Подсказка:** в Python модуль числа — это `abs(x)`

```python
def mae(predictions, true_values):
    """
    Считает среднюю абсолютную ошибку (MAE).

    predictions  -- массив предсказаний модели
    true_values  -- массив реальных значений
    """
    loss = 0
    for i in range(len(predictions)):
        loss += abs(predictions[i]-true_values[i])
    return loss/len(predictions)


# Проверка: ошибки |100-110|=10, |200-180|=20, |300-320|=20 → среднее = 50/3 ≈ 16.67
test_preds  = np.array([100, 200, 300])
test_truths = np.array([110, 180, 320])
test_mae = mae(test_preds, test_truths)
print(f"mae([100,200,300], [110,180,320]) = {test_mae:.2f}")
print(f"Ожидается: 16.67")
assert abs(test_mae - 50/3) < 0.01, "Проверьте формулу!"
print("✅ Тест пройден!")
```

### 3.2 Посчитайте MAE для своей модели

Используйте коэффициенты из Задачи 2 и посчитайте MAE на всём датасете.

```python
# Считаем предсказания для всех 120 дней сразу
predictions = predict(temperature, wind_speed, my_w_temp, my_w_wind, my_bias)

# ВАШ КОД ЗДЕСЬ: посчитайте MAE и сохраните в переменную my_mae
my_mae = mae(predictions, rides)

print(f"MAE вашей модели: {my_mae:.1f} поездок")
print(f"Среднее число поездок в датасете: {rides.mean():.0f}")
print(f"Относительная ошибка: {my_mae / rides.mean() * 100:.1f}%")
```

## Задача 3. Подбор коэффициентов вручную · *1 балл*

Теперь у нас есть инструмент оценки — MAE. Попробуем улучшить модель, меняя коэффициенты руками.

Вспомните что показали графики из Задачи 1:
- температура **положительно** влияет → `w_temp` должен быть **положительным**
- ветер **отрицательно** влияет → `w_wind` должен быть **отрицательным**

**Ваша цель:** добиться MAE **ниже 50**, меняя три числа.

> 💡 Это именно то, что делает машинное обучение — только автоматически и гораздо быстрее.

```python
# 🔧 МЕНЯЙТЕ ЭТИ ЗНАЧЕНИЯ, пока MAE не станет ниже 50
w_temp = 14.0
w_wind = -9.0
bias   = 46.0

preds = predict(temperature, wind_speed, w_temp, w_wind, bias)
loss  = mae(preds, rides)

print(f"Коэффициенты: w_temp={w_temp}, w_wind={w_wind}, bias={bias}")
print(f"MAE = {loss:.1f}")

if loss < 50:
    print("🎉 Цель достигнута!")
else:
    print(f"Пока {loss} выше цели. Попробуйте скорректировать коэффициенты.")

# Наглядный отчёт: посмотрим на первые 10 дней снова
print(f"\n{'День':>5}  {'Реально':>9}  {'Модель':>8}  {'Ошибка':>8}")
print("-" * 38)
for i in range(10):
    p = predict(temperature[i], wind_speed[i], w_temp, w_wind, bias)
    print(f"{i+1:>5}  {rides[i]:>9}  {p:>8.0f}  {abs(p - rides[i]):>8.0f}")
```

## Задача 4. Тривиальный оптимизатор · *2 балла*

Подбирать коэффициенты вручную — долго и скучно. Давайте научим **компьютер** делать это за нас!

Самый простой способ — **перебор**: попробовать много разных значений `w_temp`, `w_wind` и `bias`, посчитать MAE для каждой тройки, и запомнить ту, при которой ошибка наименьшая.

---

### 4.1 Напишите оптимизатор перебором

Реализуйте функцию `brute_force_optimizer`. Она должна:
1. Перебрать все комбинации из трёх списков значений
2. Для каждой тройки `(wt, ww, b)` посчитать MAE
3. Запомнить тройку с наименьшей ошибкой
4. Вернуть лучшую тройку и её MAE

**Подсказка:** три вложенных цикла `for`.

```python
def brute_force_optimizer(temperature, wind_speed, rides,
                          w_temp_values, w_wind_values, bias_values):
    """
    Перебирает все комбинации коэффициентов и возвращает лучшую.

    Возвращает: (best_w_temp, best_w_wind, best_bias, best_mae)
    """
    best_mae    = float('inf')  # начинаем с «бесконечной» ошибки
    best_w_temp = None
    best_w_wind = None
    best_bias   = None

    for w_temp in w_temp_values:
        for w_wind in w_wind_values:
            for bias in bias_values:
                predictions = predict(temperature, wind_speed, w_temp, w_wind, bias)
                mae_rn = mae(predictions, rides)
                if mae_rn < best_mae:
                    best_mae = mae_rn
                    best_w_temp, best_w_wind, best_bias = w_temp, w_wind, bias



    return best_w_temp, best_w_wind, best_bias, best_mae


print("Функция описана. Запустите следующую ячейку!")
```

### 4.2 Запустите оптимизатор

Мы зададим диапазоны поиска. Компьютер переберёт все комбинации — это займёт несколько секунд.

```python
w_temp_candidates = np.arange(5,  10,  1.0)   # от 5 до 25 с шагом 1
w_wind_candidates = np.arange(-20, 0,  1.0)   # от -35 до 0 с шагом 1
bias_candidates   = np.arange(50, 100, 5.0)   # от 50 до 200 с шагом 5


total = len(w_temp_candidates) * len(w_wind_candidates) * len(bias_candidates)
print(f"Всего комбинаций: {total:,}")
print("Ищем...")

best_wt, best_ww, best_b, best_loss = brute_force_optimizer(
    temperature, wind_speed, rides,
    w_temp_candidates, w_wind_candidates, bias_candidates
)

print(f"\n🏆 Лучшая тройка:")
print(f"   w_temp = {best_wt}")
print(f"   w_wind = {best_ww}")
print(f"   bias   = {best_b}")
print(f"   MAE    = {best_loss:.1f}")
print(f"\nВручную в Задаче 4 у вас было MAE = {loss:.1f}")
```

## Задача 5. Продвинутый оптимизатор · *3 балла*

Если мы хотим, чтобы наш `brute_force_optimizer` перебирал коэффициенты быстро, ему придётся делать это довольно грубо. Мы могли бы уменьшить шаг до 0.001, но тогда комбинаций стало бы в миллион раз больше и перебор шёл бы часами. Вместо этого сделаем умнее: будем искать сначала грубо, потом сужать область поиска вокруг лучшего результата и повторять с меньшим шагом — до тех пор, пока улучшение не станет совсем крошечным. Это уже довольно близко к тому, как устроены настоящие оптимизаторы в ML, так что задача не самая простая.

Что нужно сделать:

Напиши функцию `iterative_optimizer(temperature, wind_speed, rides, start_wt, start_ww, start_b, start_step, shrink_factor, min_improvement)`, которая:

* Запускает `brute_force_optimizer` в окрестности ±5 * `step` вокруг текущего лучшего решения с текущим `step`
* Если MAE улучшился больше чем на `min_improvement` — запоминает новое лучшее решение, уменьшает `step` в shrink_factor раз и повторяет
* Если улучшение меньше `min_improvement` — останавливается
На каждой итерации печатает номер итерации, текущий `step` и текущий MAE

```python
def iterative_optimizer(temperature, wind_speed, rides,
                        start_wt, start_ww, start_b,
                        start_step=1, shrink_factor=2.0, min_improvement=0.01):

    best_wt   = start_wt
    best_ww   = start_ww
    best_b    = start_b
    best_loss = mae(predict(temperature, wind_speed, best_wt, best_ww, best_b), rides)
    step      = start_step
    iteration = 1
    while True:
      wt_candidates = np.arange(best_wt - 5*step, best_wt + 5*step, step)
      ww_candidates = np.arange(best_ww - 5*step, best_ww + 5*step, step)
      b_candidates  = np.arange(best_b - 5*step, best_b + 5*step, step)
      cur_wt, cur_ww, cur_b, cur_mae = brute_force_optimizer(temperature, wind_speed, rides, wt_candidates, ww_candidates, b_candidates)
      print(iteration, step, best_loss)
      iteration += 1
      if best_loss - cur_mae > min_improvement:
        best_loss = cur_mae
        step /= shrink_factor
        best_wt, best_ww, best_b = cur_wt, cur_ww, cur_b
      else:
        break


    return best_wt, best_ww, best_b, best_loss


iter_wt, iter_ww, iter_b, iter_loss = iterative_optimizer(
    temperature, wind_speed, rides,
    start_wt=best_wt, start_ww=best_ww, start_b=best_b,
    start_step=1.0, shrink_factor=2.0, min_improvement=0.01
)
```

```python
print(f"Грубый перебор:       MAE = {best_loss:.4f}")
print(f"Итеративный:          MAE = {iter_loss:.4f}")
print()

assert iter_loss <= best_loss + 0.01, \
    "Итеративный оптимизатор не улучшил грубый перебор — проверьте логику цикла"

assert iter_loss < best_loss + 0.01, \
    "Функция вернула результат, но цикл, похоже, выполнился только один раз — проверьте условие остановки"
```

В итоге модель по сути просто подстраивается под исходные данные, то есть она не поняла закономерность, а просто стала сильнее подгоняться под обучающие данные

## 🚀 Бонус: сила настоящего ML

> *Этот раздел не оценивается — просто запустите ячейки и посмотрите, что происходит!*

Мы перебрали десятки тысяч комбинаций. Но в реальных задачах признаков бывают **сотни** — перебор занял бы годы.

Поэтому существует **машинное обучение** — алгоритмы, которые находят оптимальные параметры умно, а не перебором. Посмотрим на библиотеку `sklearn`:

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error

# Собираем признаки: каждая строка — один день, каждый столбец — один признак
X = np.column_stack([temperature, wind_speed])
y = rides

# Создаём модель
model = LinearRegression()

# Обучаем — sklearn сам найдёт лучшие коэффициенты (никакого перебора!)
model.fit(X, y)

# Предсказываем
sklearn_preds = model.predict(X)

# Считаем качество
sklearn_mae = mean_absolute_error(y, sklearn_preds)

print("=" * 50)
print("sklearn нашла:")
print(f"  w_temp = {model.coef_[0]:.2f}   (наш перебор: {best_wt})")
print(f"  w_wind = {model.coef_[1]:.2f}   (наш перебор: {best_ww})")
print(f"  bias   = {model.intercept_:.2f}   (наш перебор: {best_b})")
print(f"  MAE    = {sklearn_mae:.1f}   (наш перебор: {best_loss:.1f})")
print("=" * 50)
print(f"\nИстинные коэффициенты (из генератора):")
print(f"  w_temp={true_w_temp},  w_wind={true_w_wind},  bias={true_bias}")
```

```python
# Итоговый график сравнения — запустите и посмотрите
best_preds_bf = predict(temperature, wind_speed, best_wt, best_ww, best_b)

fig, axes = plt.subplots(1, 2, figsize=(12, 5))
for ax, preds, title, lv in zip(
    axes,
    [best_preds_bf, sklearn_preds],
    ['Наш перебор', 'sklearn (настоящий ML)'],
    [best_loss, sklearn_mae]
):
    ax.scatter(rides, preds, alpha=0.5)
    mv = max(rides.max(), max(preds))
    ax.plot([0, mv], [0, mv], 'r--', label='Идеал')
    ax.set_xlabel('Реальные поездки')
    ax.set_ylabel('Предсказанные поездки')
    ax.set_title(f'{title}\nMAE = {lv:.1f}')
    ax.legend()
    ax.grid(alpha=0.3)

plt.tight_layout()
plt.show()

print("sklearn нашла почти те же коэффициенты, что заложены в датасет —")
print("и сделала это мгновенно, без ручного перебора.")
```
